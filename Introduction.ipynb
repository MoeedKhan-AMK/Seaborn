    1. Powerful tools for visualizing data, including scatter plots, line plots, bar plots, heat maps, and many more.

- Generate attractive plots with minimal coding efforts.
- Seaborn offers a range of built-in statistical functions, allowing users to easily perform complex statistical analysis with their visualizations.

### Seaborn Plot Types
Seaborn provides a wide range of plot types that can be used for data visualization and exploratory data analysis. Broadly speaking, any visualization can fall into one of the three categories. 

    1. Univariate  x only (contains only one axis of information)
    2. Bivariate  x and y (contains two axis of information)
    3. Trivariate  x, y, z (contains three axis of information)

![alt text](Seaborn_Plot_Types_db2382dd13.png)

**Most commonly used plot types in Seaborn:**

- 1. Scatter Plot. A scatter plot is used to visualize the relationship between two variables. Seaborn's scatterplot() function provides a simple way to create scatter plots.
- 2. Line Plot. A line plot is used to visualize the trend of a variable over time. Seaborn's lineplot() function provides a simple way to create line plots.
- 3. Histogram. A histogram is used to visualize the distribution of a variable. Seaborn's histplot() function provides a simple way to create histograms.
- 4. Box Plot. A box plot is used to visualize the distribution of a variable. Seaborn's boxplot() function provides a simple way to create box plots.
- 5. Violin Plot. A violin plot is similar to a box plot, but provides a more detailed view of the distribution of the data. Seaborn's violinplot() function provides a simple way to create violin plots.
- 6. Heatmap. A heatmap is used to visualize the correlation between different variables. Seaborn's heatmap() function provides a simple way to create heatmaps.
- 7. Pairplot. A pairplot is used to visualize the relationship between multiple variables. Seaborn's pairplot() function provides a simple way to create pairplots.

```python
# sns.scatterplot()
sns.scatterplot(data= titanic, x='fare', y='age', color='Green', hue='who')
```

### `sns.set_style()`
- `sns.set_style()` is used to control the overall look/theme of your Seaborn plots (background, grid, axes).
- Common Styles:
    1. `sns.set_style("darkgrid")`   # default, very common for analysis
    2. `sns.set_style("whitegrid")`  # clean, good for reports
    3. `sns.set_style("dark")`       # dark background
    4. `sns.set_style("white")`      # minimal, no grid
    5. `sns.set_style("ticks")`     # white + ticks on axes

```python
sns.set_style('dark')
sns.scatterplot(data= titanic, x='fare', y='age', color='Green', hue='who')
```
