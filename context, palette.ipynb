# Context, Palette in Seaborn
In Seaborn, context controls the scale of plot elements ‚Äî not the colors or style.

Think of it as:
üëâ ‚ÄúHow big should everything look?‚Äù

What context affects:
- Font sizes
- Line widths
- Marker sizes
- Tick sizes

context does not affect overall style.

**There are 4 different types of context in seaborn:**
1. `'Paper'` : Smallest Elements, Ideal for printing in academic papers.
2. `'Notebook'` : Medium Size, Good for working inside Jupyter notebooks (default setting).
3. `talk` : Larger Text and Lines, Good for Presentation Slides where readability matters.
4. `'poster'` : Very Large elements, Suitable for large displays or posters.


Syntax: **`sns.set_context("notebook/paper/talk/poster")`**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Loading Dataset

titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.set_style('dark')
sns.set_context('poster')
sns.scatterplot(data= titanic, x='fare', y='age', color='Green', hue='who')
```

## Palette in Seaborn

following are some built-in pallete






```python
sns.set_style('dark')
sns.set_context('notebook')
sns.scatterplot(data= titanic, x='fare', y='age', color='Green', hue='who', palette='Set1')
```
